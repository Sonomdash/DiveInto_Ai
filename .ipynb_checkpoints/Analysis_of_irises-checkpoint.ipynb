Problem 6 Creating a diagram

Matplotlib - Violin Plot
Violin plots are similar to box plots, except that they also show the probability density of the data at different values.

Matplotlib - Box Plot
A box plot which is also known as a whisker plot displays a summary of a set of data containing the minimum, first quartile, median, third quartile, and maximum. 



```python
import matplotlib.pyplot as plt

def  column_different_ways(x,y,values):
    i=0
    array=[]
    for _ in union_x_y.index: 
        array.append(union_x_y.loc[i][values])
        i+=1
    data = []
    for index in range(len(array)):
        if array[index] >= x and array[index] < y:
            data.append(array[index])
            
    return len(data)
    
bet_2_3 = column_different_ways(2.0,3.0,1)
bet_3_4 = column_different_ways(3.0,4.0,1)
bet_4 = column_different_ways(4.0,5.0,1)
def percentages(value):
    length = len(union_x_y.index)
    val = (value *100)/length
    return val
print("                    Sepal width pie ")

labels = 'Values between 2.0 and 3.0', 'Values between 3.0 and 4.0', '4.0 max values'
sizes = [percentages(bet_2_3), percentages(bet_3_4), percentages(bet_4)]
explode = (0, 0.1, 0) 
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.



def values(x ):
    i =0
    array = []
    for _ in union_x_y.index: 
        array.append(union_x_y.loc[i][x])
        i+=1
    return array
sepal_length_data =values(0)
sepal_width_data = values(1)
petal_length_data = values(2)
petal_width_data = values(3)

data = [sepal_length_data, sepal_width_data, petal_length_data, petal_width_data]
  
fig = plt.figure(figsize =(10, 7))
  
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
  
# Creating plot
bp = ax.boxplot(data)
  
# show plot


fig = plt.figure()

# Create an axes instance
ax = fig.add_axes([0,0,1,1])

# Create the boxplot
bp = ax.violinplot(data)

```

```python
print("Create a correlation coefficient matrix for 4 features")
crr_data = pd.concat([first_x, first_y],axis=1)
def histogram_intersection(a, b):
    v = np.minimum(a, b).sum().round(decimals=1)
    return v
crr_data.corr(method=histogram_intersection)
```

Problem 8 Explaining the results of visualized graphs and tables

The first concept

Python Pandas - DataFrame
A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns.

Features of DataFrame in potentially columns are of different types  , size â€“ Mutable,  labeled axes (rows and columns), Can Perform Arithmetic operations on rows and columns


Use pandas.concat to compile multiple DataFrame


loc - lis label based data selecting method which means that we have to pass the name of the row or column which we want to select. This method includes the last element of the range passed in it, unlike iloc(). loc() can accept the boolean data unlike iloc()

.iloc -  iloc() is a indexed based selecting method which means that we have to pass integer index in the method to select specific row/column. This method does not include the last element of the range passed in it unlike loc(). iloc() does not accept the boolean data unlike loc().

matplotlib.pyplot.pie -A Pie Chart is a circular statistical plot that can display only one series of data.he area of the chart is the total percentage of the given data. The area of slices of the pie represents the percentage of the parts of the data. The slices of pie are called wedges. The area of the wedge is determined by the length of the arc of the wedge. The area of a wedge represents the relative percentage of that part with respect to whole data

matplotlib.pyplot.boxplot - A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a verticle line is also there which goes through the box at the median. Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.



matplotlib.pyplot.violinplot - Through this function, you can make a violin plot for every column of the dataset or each vector in the dataset sequence. All filled areas extend to show the entire data range with lines that are optional at the mean, the median, the maximum and the minimum

seaborn.pairplot - To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function. This shows the relationship for (n, 2) combination of variable in a DataFrame as a matrix of plots and the diagonal plots are the univariate plots.

pandas.DataFrame.corr - Pandas dataframe.corr() is used to find the pairwise correlation of all columns in the dataframe. Any na values are automatically excluded. For any non-numeric data type columns in the dataframe it is ignored.

seaborn.heatmap - Heat maps display numeric tabular data where the cells are colored depending upon the contained value. Heat maps are great for making trends in this kind of data more readily apparent, particularly when the data is ordered and there is clustering.
